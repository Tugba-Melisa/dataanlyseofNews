```python
import matplotlib.pyplot as plt

# Veri hazırlığı
categories = list(data.keys())
words = list(data[categories[0]].keys())
probabilities = {word: [data[category][word] for category in categories] for word in words}

# Çubuk grafik oluşturma
fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.15
opacity = 0.8

for i, (word, probs) in enumerate(probabilities.items()):
    positions = [j + bar_width * i for j in range(len(categories))]
    plt.bar(positions, probs, bar_width, alpha=opacity, label=word)

# Eksen etiketleri ve başlık
plt.xlabel('Kategoriler')
plt.ylabel('Olasılık Değerleri')
plt.title('Konulara Göre Kelime Olasılıkları')
plt.xticks([r + bar_width * (len(words) / 2) for r in range(len(categories))], categories)
plt.legend()

# Grafiği göster
plt.tight_layout()
plt.show()
```
